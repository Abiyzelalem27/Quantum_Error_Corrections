### Defining a channel and  Bloch Sphere Visualization of Quantum Noise Channels

Generate an array of Kraus operators $M_i$ for the phase damping channel.
Write a function that takes as an input (1) a density matrix of a single qubit and (2) the list of Kraus operators defining the channels and returns the density matrix after the channel has been applied.

Test your functions by applying, e.g., the to the states $|0\rangle$ and $|+\rangle$.
Phase Damping channels affect the qubit states |0⟩ and |+⟩. Green arrows represent the original states, and red arrows represent the states after the noise channel.

```python
lmbda = 0.9
psi = qec.operators.ket0()
rho_in =  qec.operators.dm(psi)

print("psi:\n", psi )
print()
print("rho_in:\n", rho_in)

kraus_phase_damping =  qec.operators.phase_damping_kraus(lmbda )
print()
print("Phase_damping_kraus:\n", kraus_phase_damping)
print()
rho_out =  qec.operators.apply_kraus(rho_in, kraus_phase_damping)

print("Density matrix after phase damping:\n", rho_out)
print()
v_initial =  qec.operators.bloch_vector(rho_in)
v_final =  qec.operators.bloch_vector(rho_out)

print("Initial Bloch vector:\n", v_initial)
print()
print("After phase_damping:\n", v_final)

b = Bloch(figsize=(3,3))
b.vector_color = ['green','red']
b.add_vectors(v_initial)
b.add_vectors(v_final)
b.title = "Phase_damping Channel"
b.show()
```

### Fidelity under Phase-Damping Noise

**phase-damping channel** models a type of decoherence where the relative phase of a qubit is disturbed without energy loss.  

- For the **\(|0\rangle\) or \(|1\rangle\) states**, the phase is unchanged, so the fidelity remains 1.  
- For superposition states like **\(|+\rangle\)**, the phase damping reduces coherence, lowering the fidelity as the decoherence probability increases.  

The **fidelity** quantifies how close the final state is to the original state.  

Thus, the **fidelity decreases with increasing phase-damping probability**, and a plot of fidelity vs. decoherence probability illustrates this behavior.

```python
psi = qec.operators.ket_plus()
rho = qec.operators.rho_from_state(psi)

lams = np.linspace(0, 1, 51)
F = []

for lam in lams:
    rho_out =qec.operators. apply_kraus(rho, qec.operators.phase_damping_kraus(lam))
    F.append(qec.operators.fidelity_pure_vs_rho(psi, rho_out))
    
lam_test = 0.5
rho_test = qec.operators.apply_kraus(rho, qec.operators.phase_damping_kraus(lam_test))
print("lambda =", lam_test)
print("rho_out =\n", rho_test)

# Plot
plt.figure(figsize=(7,4))
plt.plot(lams, F, marker="o")
plt.xlabel("λ")
plt.ylabel("Fidelity")
plt.title("Fidelity vs λ")
plt.grid(True)
plt.show()
```
