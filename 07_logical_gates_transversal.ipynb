 07 â€” Transversal Logical Gates (Bit-Flip Code)

In quantum error correction, some logical operations can be applied transversally.
That means we apply a gate independently on each physical qubit, which is good because it does not spread errors.

---

3-qubit bit-flip code

Encoding:

\[
|\psi\rangle = \alpha|0\rangle + \beta|1\rangle
\quad\Rightarrow\quad
|\psi_L\rangle = \alpha|000\rangle + \beta|111\rangle.
\]

The logical \(X\) operator is:

\[
X_L = X_1 X_2 X_3.
\]

We verify by simulation that:

\alpha|111\rangle + \beta|000\rangle.
\]


```python
import numpy as np

def kron(*ops):
    """Kronecker product of many operators."""
    out = np.array([[1]], dtype=complex)
    for op in ops:
        out = np.kron(out, op)
    return out

def encode_bitflip3(alpha, beta):
    """|psi> -> alpha|000> + beta|111> (basis order |q2 q1 q0>)"""
    psi_L = np.zeros(8, dtype=complex)
    psi_L[0] = alpha
    psi_L[7] = beta
    return psi_L

# Logical X gate (transversal)
X_L = kron(X, X, X)

# Example encoded state
alpha = 0.6
beta = np.sqrt(1 - alpha**2)

psi_L = encode_bitflip3(alpha, beta)
psi_out = X_L @ psi_L

# Expected result: alpha|111> + beta|000>
expected = np.zeros(8, dtype=complex)
expected[7] = alpha
expected[0] = beta

print("Output matches expected:", np.allclose(psi_out, expected))
print("psi_out =", psi_out)
```
