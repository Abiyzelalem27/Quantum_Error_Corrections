```python
import numpy as np

# Apply single-qubit error
def apply_error(state, error, index, num_qubits):
    """Apply X, Y, Z error on a single qubit (0-indexed)"""
    if error == 'I':
        return state
    pauli_dict = {'X':X, 'Y':Y, 'Z':Z}
    op = 1
    for i in range(num_qubits):
        if i == index:
            op = np.kron(op, pauli_dict[error])
        else:
            op = np.kron(op, I)
    return op @ state


# Detect Errors
def detect_errors(state):
    """Return +1/-1 outcomes for bit and phase stabilizers"""
    bit_results = [1 if np.isclose(measure_stabilizer(state,s), 1) else -1 for s in bit_stabilizers]
    phase_results = [1 if np.isclose(measure_stabilizer(state,s),1) else -1 for s in phase_stabilizers]
    return bit_results, phase_results


# Correct Errors
def correct_errors(state):
    """Detects error and applies correction"""
    bit_results, phase_results = detect_errors(state)
    
    # Bit-flip correction: identify qubit in block
    for i in range(3):
        if bit_results[i]==-1 and bit_results[i+1]==-1:
            qubit_to_flip = i+1  # qubit index 0-based
            state = apply_error(state,'X', qubit_to_flip, 9)
            break
    # Phase-flip correction: identify block
    if phase_results[0]==-1 and phase_results[1]==+1:
        # Block 1 phase error
        state = apply_error(state,'Z',0, 9)
    elif phase_results[0]==-1 and phase_results[1]==-1:
        # Block 2 phase error
        state = apply_error(state,'Z',3, 9)
    elif phase_results[0]==+1 and phase_results[1]==-1:
        # Block 3 phase error
        state = apply_error(state,'Z',6, 9)
    
    return state

# Example: Detect & Correct Z1 on logical |0>_L

state = shor_logical_zero()
state_error = apply_error(state, 'Z', 4, 9)  # Apply Z on qubit 0 (Z1) 
bit_res, phase_res = detect_errors(state_error)
print("Bit-flip stabilizers:", bit_res)
print("Phase-flip stabilizers:", phase_res)

state_corrected = correct_errors(state_error)
bit_res_corr, phase_res_corr = detect_errors(state_corrected)
print("After correction - Bit-flip:", bit_res_corr)
print("After correction - Phase-flip:", phase_res_corr)
```
