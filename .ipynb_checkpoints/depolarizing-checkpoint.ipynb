### Bloch Sphere Visualization of Quantum Noise Channels

Depolarizing channels affect the qubit states \(|0\rangle\) and \(|+\rangle\). Green arrows represent the original states, and red arrows represent the states after the noise.

Generate an array of Kraus operators \(M_i\) for a depolarizing channel.

Write a function that takes as input a single-qubit density matrix and a list of Kraus operators, and returns the density matrix after the channel is applied.

Test your functions by applying the channel to the states \(|0\rangle\) and \(|+\rangle\).

```python
p = 0.5
psi = qec.operators.ket0() 
rho_in = qec.operators.dm(psi)
print("psi:\n", psi )
print()
print("rho_in:\n", rho_in)
kraus_dep = qec.operators.depolarizing_kraus(p)
print()
print("Kraus operators for depolarizing channel:\n", kraus_dep)
rho_out = qec.operators.apply_kraus(rho_in, kraus_dep)
print()
print("Density matrix after depolarizing channel:\n", rho_out)
v_initial = qec.operators.bloch_vector(rho_in)
v_final = qec.operators.bloch_vector(rho_out)
print()
print("Initial Bloch vector:\n", v_initial)
print()
print("Bloch vector after depolarizing channel:\n", v_final)

b = Bloch()
b.vector_color = ['green','red']
b.add_vectors(v_initial)
b.add_vectors(v_final)
b.title = "Depolarizing Channel"
b.show()
```

### n-bit Register: Testing the Depolarizing Channel

Test the implementation on a multi-qubit system where the expected result is known. For example, apply the depolarizing channel to the \(|0\rangle\) state for \(n = 4\).

```python
p = 0.2
n = 4
psi = qec.operators.ket0
initial_state = np.kron(np.kron(psi(), psi()), np.kron(psi(), psi()))
print("initial_state :\n", initial_state) 
print()
rho_in = np.outer(initial_state, initial_state.conj())
kraus_dep = qec.operators.depolarizing_kraus_nqubits(p, n)
rho_out = qec.operators.apply_kraus(rho_in, kraus_dep)
print("Density matrix:\n", rho_in)
print()
print("Density matrix after depolarizing channel:\n", rho_out)
```

```python
p = 0.2
n = 4
psi = qec.operators.ket_plus
initial_state = np.kron(np.kron(psi(), psi()), np.kron(psi(), psi()))
print("initial_state :\n", initial_state) 
print()
rho_in = np.outer(initial_state, initial_state.conj())
kraus_dep = qec.operators.depolarizing_kraus_nqubits(p, n)
rho_out = qec.operators.apply_kraus(rho_in, kraus_dep)
print("Density matrix:\n", rho_in)
print()
print("Density matrix after depolarizing channel:\n", rho_out)
```

### Fidelity in a Depolarizing Channel

A **depolarizing channel** models noise where a qubit randomly becomes a completely mixed state with a certain probability \(p\).  

- The **fidelity** measures how close the final noisy state is to the original state.  
- For a pure state, it is calculated as the overlap between the original state and the noisy state.  

As the **depolarizing probability \(p\) increases**, the fidelity **decreases**, showing the increasing deviation of the qubit from its original state.  


```python
#initial state
alpha = 0.6
beta = np.sqrt(1 - alpha**2)
psi = np.array([alpha, beta], dtype=complex)
psi = qec.operators.normalize_state(psi)
psi = np.array([[1], [0]], dtype=complex)
rho = psi @ psi.conj().T

# Sweep over depolarizing probability p
p_vals = np.linspace(0, 1, 50)
fid_vals = []

for p in p_vals:
    kraus = qec.operators.depolarizing_kraus(p)
    rho_out = qec.operators.apply_kraus(rho, kraus)
    fid_vals.append(qec.operators.fidelity_pure(psi, rho_out))

# Analytical fidelity formula for comparison
fid_analytical = 1 - 2*p_vals/3

# Plot numerical and analytical fidelities
plt.figure(figsize=(7,5))
plt.plot(p_vals, fid_vals, label="Numerical", marker='o', markersize=3)
plt.plot(p_vals, fid_analytical, '--', label="Analytical", color='red')
plt.xlabel("Probability")
plt.ylabel("Fidelity")
plt.title("Fidelity vs Probability")
plt.grid(True)
plt.legend()
plt.show()
```
