### Bloch Sphere Visualization of Quantum Noise Channels

Phase flip channels affect the qubit states |0⟩ and |+⟩. Green arrows represent the original states, and red arrows represent the states after the noise channel.

```python
theta = np.pi / 4  
p = 0.1            
n = [0, 0, 1]    
psi = qec.operators.ket0() 
rho_in = np.outer(psi, psi.conj())  
print("psi:\n", psi )
print()
print("rho_in:\n", rho_in)
R_z  = qec.operators.rotation_gate(theta, n)
print()
print("Z-axis:\n", R_z)
print()
kraus_z = qec.operators.rotation_channel(p, R_z)
rho_z = qec.operators.apply_kraus(rho_in, kraus_z)
print("Kraus operators for Z-axis channel:\n", kraus_z)
print()
print("Density matrix after Z channel:\n", rho_z)
print()
v_initial = qec.operators.bloch_vector(rho_in)
v_z = qec.operators.bloch_vector(rho_z)
print("initial Bloch vector:", v_initial)
print("Bloch vector after Z channel:", v_z)
#Bloch Visulations 
b = Bloch(figsize=(3,3))
b.vector_color = ['green', 'red']  
b.add_vectors(v_initial)  
b.add_vectors(v_z)        
b.title = "Z-axis effect on |0>"
b.show()
```

### Fidelity of a Qubit under Phase-Flip Noise

phase-flip channel introduces errors by flipping the phase of a qubit with probability \(p\).  

The fidelity measures how close the noisy state is to the original state.  
This means the fidelity decreases linearly with the phase-flip probability, and a plot of fidelity vs. \(p\) clearly illustrates this trend.

```python
# Fidelity for Phase-Flip channel |+> state
def fidelity_phaseflip_plus(p):
    # Fidelity decreases linearly: F = 1 - p
    return 1 - p

# Probabilities from 0 to 1
p_vals = np.linspace(0, 1, 50)

# Compute fidelities
fid_phaseflip = fidelity_phaseflip_plus(p_vals)

# Specific value p = 0.5
p_spec = 0.5
fid_phaseflip_spec = fidelity_phaseflip_plus(np.array([p_spec]))[0]

# Plot fidelity
plt.figure(figsize=(8,6))
plt.plot(p_vals, fid_phaseflip, 'g-', linewidth=2, label='Phase-Flip Channel |+>')
plt.scatter(p_spec, fid_phaseflip_spec, color='green', s=100, zorder=5)
plt.text(p_spec+0.02, fid_phaseflip_spec, f'{fid_phaseflip_spec:.2f}', color='green', fontsize=12)

plt.xlabel("Probability")
plt.ylabel("Fidelity")
plt.title("Fidelity vs Probability")
plt.grid(True)
plt.legend()
plt.show()
```
